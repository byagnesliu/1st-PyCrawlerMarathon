```python
import xmltodict

with open("./Data/63_72hr_CH.xml","r",encoding="utf-8") as xmlfile:
    content=dict(xmltodict.parse(xmlfile.read()))
    print(content)
```

* 請問高雄市有多少地區有溫度資料？

```python
print('高雄市裡共有', len(content['cwbopendata']['dataset']['locations']['location']), '個地區有溫度資料，分別是:\n')
for i, place in enumerate(content['cwbopendata']['dataset']['locations']['location']):
    print(i+1 , place['locationName'])
```

* 請取出每一個地區所記錄的第一個時間點跟溫度

```python
for k in content['cwbopendata']['dataset']['locations']['location']:
    print(k['locationName'], '時間', k['weatherElement'][0]['time'][0]['dataTime'], '溫度',k['weatherElement'][0]['time'][0]['elementValue']['value'], '°C' )
```

* 請取出第一個地區所記錄的每一個時間點跟溫度

```python
print(content['cwbopendata']['dataset']['locations']['location'][0]['locationName'], '的氣溫量測數據:\n')
for k in content['cwbopendata']['dataset']['locations']['location'][0]['weatherElement'][0]['time']:
    print('時間:', k['dataTime'], '溫度', k['elementValue']['value'], '°C')
```
