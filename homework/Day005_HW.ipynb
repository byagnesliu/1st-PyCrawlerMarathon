## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
import requests
res=requests.get("https://www.dcard.tw/_api/forums/pet/posts?popular=true")
res.text
```

```python
import json
comments=json.loads(res.text)

print("一次會回傳幾筆資料：", len(comments),"筆資料")
print("每一筆資料包含欄位：\n",comments[0].keys())
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
count = 1
for data in comments:
    print(str(count)+'.\t'+'標題:{0:10s} | 貼文時間:{1:25s} | 留言人數{2:4d} | 按讚人數{3:4d}'.format(data['title'], data['createdAt'], data['commentCount'], data['likeCount']))
    count=count+1
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
res_pop = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
res_npop = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')

pop_datas=json.loads(res_pop.text)
pop_i=0
comment_count=0
like_count=0
for pop_data in pop_datas:
    comment_count+=pop_data["commentCount"]
    like_count+=pop_data["likeCount"]
    pop_i+=1

print("==熱門文章==\n平均留言人數:"+str(comment_count/pop_i)+"平均按讚人數:"+str(like_count/pop_i))

npop_datas=json.loads(res_npop.text)
npop_i=0
comment_count_npop=0
like_count_npop=0
for npop_data in npop_datas:
    comment_count_npop+=npop_data["commentCount"]
    like_count_npop+=npop_data["likeCount"]
    npop_i+=1

print("==非熱門文章== \n平均留言人數:"+str(comment_count_npop/npop_i)+"\n平均按讚人數:"+str(like_count_npop/npop_i))
```
