```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ
import requests
res=requests.get("https://www.dcard.tw/_api/forums/pet/posts?popular=true")
res.text
```

```python
import json
comments=json.loads(res.text)

print("ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼š", len(comments),"ç­†è³‡æ–™")
print("æ¯ä¸€ç­†è³‡æ–™åŒ…å«æ¬„ä½ï¼š\n",comments[0].keys())
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€
count = 1
for data in comments:
    print(str(count)+'.\t'+'æ¨™é¡Œ:{0:10s} | è²¼æ–‡æ™‚é–“:{1:25s} | ç•™è¨€äººæ•¸{2:4d} | æŒ‰è®šäººæ•¸{3:4d}'.format(data['title'], data['createdAt'], data['commentCount'], data['likeCount']))
    count=count+1
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
res_pop = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
res_npop = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')

pop_datas=json.loads(res_pop.text)
pop_i=0
comment_count=0
like_count=0
for pop_data in pop_datas:
    comment_count+=pop_data["commentCount"]
    like_count+=pop_data["likeCount"]
    pop_i+=1

print("==ç†±é–€æ–‡ç« ==\nå¹³å‡ç•™è¨€äººæ•¸:"+str(comment_count/pop_i)+"å¹³å‡æŒ‰è®šäººæ•¸:"+str(like_count/pop_i))

npop_datas=json.loads(res_npop.text)
npop_i=0
comment_count_npop=0
like_count_npop=0
for npop_data in npop_datas:
    comment_count_npop+=npop_data["commentCount"]
    like_count_npop+=npop_data["likeCount"]
    npop_i+=1

print("==éç†±é–€æ–‡ç« == \nå¹³å‡ç•™è¨€äººæ•¸:"+str(comment_count_npop/npop_i)+"\nå¹³å‡æŒ‰è®šäººæ•¸:"+str(like_count_npop/npop_i))
```
